Q3.a.which region has the highest GDP per capita (current US$)
 

```python
Q3a = dataw6.groupby("Region")['GDP per capita (current US$)'].mean().round().reset_index()
Q3a
```

```python
Q3a.nlargest(1,'GDP per capita (current US$)')
```

Ans. NorthernAmerica tops in GDP

Q3b. which region has the highest International trade: Imports (million US$)

```python
Q3b = dataw6.groupby("Region")['International trade: Imports (million US$)'].mean().round().reset_index()
Q3b
```

```python
Q3b.nlargest(1,'International trade: Imports (million US$)')
```

Ans. NorthernAmerica tops in International trade: Imports (million US$)
